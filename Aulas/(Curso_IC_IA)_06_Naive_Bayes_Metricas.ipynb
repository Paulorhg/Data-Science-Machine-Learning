```python
import pandas as pd

df = pd.read_csv('AUS_cobar.csv')
df = df[['MinTemp','MaxTemp','Humidity3pm','RainTomorrow']]
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['MinTemp','MaxTemp','Humidity3pm'])
```

```python
X = df[['MinTemp','MaxTemp','Humidity3pm']]
y = df.RainTomorrow
```

```python
df.head(15)
```

```python
X_train, X_test = X[:-300], X[-300:]
y_train, y_test = y[:-300], y[-300:]
```

```python
from sklearn.neighbors import KNeighborsClassifier

modelo = KNeighborsClassifier(n_neighbors = 1)
modelo.fit(X_train, y_train)
y_pred = modelo.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_pred, y_test)

from sklearn.metrics import classification_report

print(classification_report(y_pred, y_test))
```

```python
X_train.describe()
```

```python
from sklearn.naive_bayes import BernoulliNB

modelo = BernoulliNB(binarize=70)
modelo.fit(X_train, y_train)
y_pred = modelo.predict(X_test)
```

```python
accuracy_score(y_pred, y_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_pred, y_test))
```

```python
from sklearn.model_selection import GridSearchCV

modelo = BernoulliNB()
params = {'binarize':[12,18,26,30]}
clf = GridSearchCV(modelo, params)
clf.fit(X_train, y_train)
clf_pred = clf.predict(X_test)
print(classification_report(clf_pred, y_test))
```

```python
display(pd.DataFrame(clf.cv_results_))
```
