```python
df['RainToday'] = df.RainToday.map({"No":0, "Yes":1})
df['WindGustDir'] = df['WindGustDir'].str[:1]

df_remove = df.dropna(subset=['MinTemp', 'MaxTemp', 'WindGustSpeed', 'Pressure9am', 'Pressure3pm','Humidity3pm',
                              'Humidity9am', 'Temp9am', 'Temp3pm', 'RainToday', 'WindGustDir'])

```

```python
#x = df.drop(columns=['RISK_MM', 'RainTomorrow'])
x = df_remove[['MinTemp', 'MaxTemp', 'WindGustSpeed', 'Pressure9am', 
               'Pressure3pm','Humidity3pm', 'Humidity9am', 'Temp9am', 'Temp3pm', 'RainToday']]

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)

y = df_remove.RainTomorrow
```

```python
pd.DataFrame(x).describe()
```

```python
x.shape, y.shape
```

```python
limite = int(len(x)* 0.7)
x_train, x_test = x[:limite], x[limite:]
y_train, y_test = y[:limite], y[limite:]
```

```python
from sklearn.neighbors import KNeighborsClassifier

modelo = KNeighborsClassifier(n_neighbors=6)
modelo.fit(x_train, y_train)
```

```python
y_pred = modelo.predict(x_test)
```

```python
#y_pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_pred, y_test)
```
